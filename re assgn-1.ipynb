```python
url='https://www.espncricinfo.com/series/ipl-2021-1249214/points-table-standings'
```

```python
a=requests.get(url)
a.text
```

```python
a
```

```python
b=a.content
```

```python
soup=BeautifulSoup(b,'html.parser')
```

```python
print(soup.prettify())
```

```python
title=soup.find('title')
title.text
```

```python
score_table=soup.find('tbody')
score_table
```

```python
print(score_table.prettify())
```

```python
teams=[]
for listing in score_table.find_all('tr'):
    print(listing.prettify())
```

```python
a=score_table.find_all('tr', class_="ds-border-b ds-border-line ds-text-tight-s")
a
```

```python
b=score_table.find_all('td', class_="ds-min-w-max ds-border-r ds-border-line ds-bg-fill-content-prime")
b
```

```python
c=score_table.find_all('div', class_="ds-flex ds-items-center ds-space-x-2")
c
```

```python
d=score_table.find_all('span', class_="ds-text-tight-s ds-font-bold ds-uppercase ds-text-left")
d
```

```python
teams=[]
for x in d:
    teams.append(x.text)
teams
```

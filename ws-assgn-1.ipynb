```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url='http://system-controls.com/industry/aerospace-3/'
```

```python
a=requests.get(url)
```

```python
a
```

```python
a.text
```

```python
b=a.content
```

```python
soup=BeautifulSoup(b,'html.parser')
```

```python
print(soup.prettify())
```

```python
title=soup.find('title')
title.text
```

```python
p_n=soup.find('body')
p_n
```

```python
print(p_n.prettify())
```

```python
a=p_n.find_all('div',class_='column col-md-8 col-sm-6 col-xs-12 content')
a                    
```

```python
b=p_n.find_all('ol', class_='key-points')
b
```

```python
c=p_n.find_all('a', class_='textclr')
c
```

```python
projects=[]
for x in c:
    projects.append(x.text)
projects
```

```python
import pandas as pd
```

```python
dict={'Aerospace Projects':projects}
```

```python
print(dict)
```

```python
bb=pd.DataFrame(dict)
```

```python
bb
```
